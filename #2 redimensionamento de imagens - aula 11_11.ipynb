```python
import cv2 
import numpy as np
from matplotlib import pyplot as plt

#carregando a imagem
caminho_para_lena = 'data-input/lena.png'
imagem_lena = cv2.imread(caminho_para_lena)
```

```python
#importando as bibliotecas


print(f'Altura em pixels: {imagem_lena.shape[0]} \n Largura em pixels: {imagem_lena.shape[1]} \n Quantidade de canais: {imagem_lena.shape[2]} ')
```

```python
for y in range(0, imagem_lena.shape[0]):
    for x in range(0, imagem_lena.shape[1]):
        imagem_lena[y,x] = (255,0,0)

cv2.imshow('Window',imagem_lena)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
for y in range(0, imagem_lena.shape[0], 20):
    for x in range(0, imagem_lena.shape[1], 20):
        imagem_lena[y:y+10, x: x+10] = (255,0,0)

cv2.imshow('Modified Image',imagem_lena)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
recorte = imagem_lena[10:200,10:200]
cv2.imshow('Recorte de Imagem',recorte)
cv2.waitKey(0)
cv2.destroyAllWindows()

cv2.imwrite("temp/recorte.jpg",recorte)
```

```python
(alt,lar) = imagem_lena.shape[:2]
centro = (lar // 2, alt // 2)

M = cv2.getRotationMatrix2D(centro, 60, 1.0)
img_rotacionada = cv2.warpAffine(imagem_lena, M, (lar,alt))
cv2.imshow('Imagem rotacionada em 60 graus', img_rotacionada)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
largura = imagem_lena.shape[1]
altura = imagem_lena.shape[0]
canais = imagem_lena[2]

proporcao = float(altura/largura)
largura_nova = 300
altura_nova = int(largura_nova*proporcao)
tamanho_novo = (largura_nova, altura_nova)

img_redimensionada = cv2.resize(imagem_lena, tamanho_novo, interpolation = cv2.INTER_AREA)
cv2.imshow('Resultado', img_redimensionada)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
