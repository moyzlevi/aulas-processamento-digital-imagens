```python
import cv2

cinza = (28, 40, 51)
caminho_para_lena = '../lena.png'
imagem_lena = cv2.imread(caminho_para_lena)

print(imagem_lena.shape[0])
```

```python
# Pintando a imagem de azul
imagem_lena[:15,:25] = cinza

cv2.imshow('img lena',imagem_lena)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Redimensionando a imagem
altura,largura = imagem_lena.shape[:2]
proporcao = float(altura/largura)
largura_nova = 200
altura_nova = int(largura_nova*proporcao)

tamanho_novo = (largura_nova, altura_nova)

img_redimensionada = cv2.resize(imagem_lena, tamanho_novo, interpolation=cv2.INTER_AREA)
cv2.imshow('Result', img_redimensionada)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Cropando a imagem
imagem_lena2 = cv2.imread('../lena.png')
recorte_lena = imagem_lena2[20:60,20:60]
cv2.imshow("Recorte de imagem", recorte_lena)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
(alt,lar) = imagem_lena2.shape[:2]
centro = (alt//2, lar//2)
matriz_transformacao = cv2.getRotationMatrix2D(centro,180,1.0)
img_rotacionada = cv2.warpAffine(imagem_lena2,matriz_transformacao,(lar,alt))

cv2.imshow('Img rotacionada',img_rotacionada)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
